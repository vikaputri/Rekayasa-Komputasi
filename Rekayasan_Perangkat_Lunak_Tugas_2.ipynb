**Menghitung Nilai d**

```python
import math

def akarkuadrat(a, b, c): 
    d = b * b - 4 * a * c 
    akar = math.sqrt(abs(d)) 
      
    if d > 0: 
        print('akar nyata dan berbeda, yaitu', (-b + akar)/(2 * a)), 'dan', ((-b - akar)/(2 * a))
      
    elif d == 0: 
        print('akar nyata dan sama, yaitu', (-b)/(2 * a))
      
    else:
        print("Akar Kompleks, yaitu") 
        print(- b / (2 * a), akar, " + i") 
        print(- b / (2 * a), akar, " - i") 
```

# Persamaan Kubik

**1.   Metode Biseksi**

```python
def bisection(f,a,b,N):
    if f(a)*f(b) >= 0:
        print("Metode bisection gagal.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        m_n = (a_n + b_n)/2
        f_m_n = f(m_n)
        if f(a_n)*f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n)*f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Menemukan solusi yang tepat.")
            return m_n
        else:
            print("Metode bisection gagal.")
            return None
    return (a_n + b_n)/2
```

```python
f = lambda x: x**2 - x - 1
approx_phi = bisection(f,1,2,25)
print(approx_phi)
```

**2.   Metode Newton-Raphson**

```python
def newton(f,Df,x0,epsilon,max_iter):
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Solusi ditemukan setelah',n,'iterasi.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Turunan nol. Tidak ada solusi yang ditemukan.')
            return None
        xn = xn - fxn/Dfxn
    print('Melebihi iterasi maksimum. Tidak ada solusi yang ditemukan.')
    return None
```

```python
f = lambda x: x**2 - x - 1
Df = lambda x: 2*x - 1
newton(f,Df,1,1e-8,10)
```

**3.   Metode Sekan**

```python
def sekan(f,a,b,N):
    if f(a)*f(b) >= 0:
        print("Metode Sekan Gagal.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        m_n = a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
        f_m_n = f(m_n)
        if f(a_n)*f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n)*f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Menemukan solusi yang tepat.")
            return m_n
        else:
            print("Metode Sekan Gagal.")
            return None
    return a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
```

```python
p = lambda x: x**3 - x**2 - 1
hasil = sekan(p,1,2,20)
print(hasil)
```
